```
from sympy import *
x=Symbol("x")
sum=0
n=4
for i in range(1,n+1):
  sum +=i
c=1/sum
def f(x):
  return c*x

print("c) C DEÄžERÄ° ")
print("c = 1/Î£f(x) => c = {}\n\n".format(c))

print("d) E(X) DEÄžERÄ°")
print("E(x) = Î£xf(x) => ")
sum = 0
str_=""
for i in range(1,n+1):
  print("{}*{} => {}".format(i,round(f(i),4),round(i*f(i),4)))
  sum +=round(i*f(i),4)
  if i==1:
    str_="{}*{}".format(i,round(f(i),4))
  else:
     str_ +="+{}*{}".format(i,round(f(i),4))
    
print("\n\n{} => E(x) = {}\n".format(str_,sum))
ex=sum**2
sum = 0

print("e) Var(x) DEÄžERÄ°")
str_=""
for i in range(1,n+1):
  print("{}^2*{} => {}".format(i,round(f(i),4),round(i**2*f(i),4)))
  sum +=round(i**2*f(i),4)
  if i==1:
    str_="{}*{}".format(i,round(f(i),4))
  else:
     str_ +="+{}*{}".format(i,round(f(i),4))

print("\n var(x) = E(x^2)-[E(x)]^2 => {}-{} => {}".format(sum,ex,sum-ex))

print("e) P(x=1)) DEÄžERÄ°")
print("\n P(x=1) => {} ".format(f(1)))

print("e) P(2<x<=4)  DEÄžERÄ°")
print("\n P(2<x<=4) => Î£f(x) = f(3)+f(4) =>  {} + {} => {} ".format(round(f(3),4),round(f(4),4),round(f(3)+f(4),4)))

print("e) P(x<=3) DEÄžERÄ°")

print("\n P(x<=3) => 1-f(4)  => 1- {} => {} ".format(round(f(4),4),round(1-f(4),4)))




```
